```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 指定 Google Drive 上的字體路徑
font_path = "/content/drive/My Drive/相關路徑/taipei_sans_tc_beta.ttf"
my_font = fm.FontProperties(fname=font_path)

plt.plot([1, 2, 3], [4, 5, 6])
plt.title("測試字體", fontproperties=my_font)
plt.show()
```
